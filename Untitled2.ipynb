```python
BAC=  yf.download('BAC',start,end)
C=  yf.download('C',start,end)
GS=  yf.download('GS',start,end)
JPM=  yf.download('JPM',start,end)
MS=  yf.download('MS',start,end)
WFC=  yf.download('WFC',start,end)
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']


print(tickers)
```

```python
###we combine dataframes into one big one
```

```python

bank_stocks  = pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
bank_stocks.head()
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info','']
```

```python
bank_stocks.head()
```

```python
bank_stocks.xs(key='Close',level='Stock Info',axis=1).max()
```

```python
returns = pd.DataFrame()
print(returns)
```

```python
for tick in tickers:
    returns[tick+'Return']=bank_stocks[tick]['Close'].pct_change()
```

```python
## the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**
fig, ax = plt.subplots(figsize=(10, 5))

# Rysowanie wykresu na tej samej osi
BAC['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(ax=ax, label='30-day mov avg')
BAC['Close'].loc['2008-01-01':'2009-01-01'].plot(ax=ax, label='BAC Close')
# Dodanie legendy i tytułu
plt.legend()
plt.title("BAC Stock Price and 30-day Moving Average")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```
